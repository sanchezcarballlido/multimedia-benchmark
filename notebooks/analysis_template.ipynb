# Results Analysis Template

Use this notebook to load and visualize the data generated by the framework.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

## 1. Load Data

Point to the path of the `combined_data.csv` file generated in your results folder.

```python
# Replace this path with the path to your results file
results_path = '../results/template_experiment_2025-06-13_14-30-00/combined_data.csv'

df = pd.read_csv(results_path)
df.head()
```

## 2. Visualization: RD (Rate-Distortion) Plot

Example of a plot comparing Bitrate vs. VMAF for different codecs.

```python
plt.figure(figsize=(12, 7))

sns.lineplot(
    data=df,
    x='bitrate_kbps',
    y='vmaf',
    hue='codec',
    style='preset',
    marker='o',
    markersize=8
)

plt.title('Rate-Distortion Curve (VMAF vs. Bitrate)')
plt.xlabel('Bitrate (kbps)')
plt.ylabel('VMAF Score')
plt.legend(title='Configuration')
plt.grid(True, which='both', linestyle='--')
plt.show()
```
